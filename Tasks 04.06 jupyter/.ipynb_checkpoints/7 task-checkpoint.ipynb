```python
#7) Create a box plot to visualize the distribution of a variable across different categories.

import pandas as pd
data = pd.read_excel(r'C:\Users\pauli\Desktop\PhytonCourse\healthy_year.xlsx')
data
```

```python
import matplotlib.pyplot as pl
years = ['2017','2018','2019','2020']
y_2017 = data['2017']
y_2018 = data['2018']
y_2019 = data['2019']
y_2020 = data['2020']

data1 = [y_2017, y_2018, y_2019, y_2020]

pl.boxplot(data1, labels=years)

pl.title("Average of healthy years by females and males")
pl.xlabel("Year")
pl.ylabel("Av_years")

pl.show()
```
