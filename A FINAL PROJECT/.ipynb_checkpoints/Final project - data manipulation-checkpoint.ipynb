```python
import pandas
data = pandas.read_csv("https://data.sfgov.org/api/views/rkru-6vcg/rows.csv?accessType=DOWNLOAD")
data
```

```python
data.to_csv('Data_orginal.csv')
```

```python
data1 = data.drop_duplicates()
data1
```

```python
data1.to_csv('Data without duplikates.csv')
```

```python
data1.dtypes
```

```python
# Filter columns
data2=data1.filter(items=['GEO Region','GEO Summary','Passenger Count'])
print(data2)
```

```python
data3 = data2.groupby(["GEO Summary", "GEO Region"]).sum()
data3
```

```python
data1['Activity Period'] = data1['Activity Period'].astype('string')
data1.dtypes
```

```python
# # make the new columns using string indexing
data1['Year'] = data1['Activity Period'].str[0:4]
data1['Month'] = data1['Activity Period'].str[4:6]


# # get rid of the extra variable (if you want)
data1.drop('Activity Period', axis=1, inplace=True)
data1
```

```python
data1.isna().any()
```

```python
# data.drop(['Operating Airline IATA Code', 'Published Airline IATA Code'], axis=1, inplace=True)
data5=data1.drop(['Operating Airline IATA Code','Published Airline','Published Airline IATA Code','Price Category Code','Terminal','Boarding Area'], axis=1)
data5
```

```python
data5.to_csv('Data transformated.csv')
```

```python
result = data5.loc[data1['Year'] == '2005', 'Passenger Count'].sum()
result
```

```python
data6 = data5.pivot_table(values='Passenger Count',index=['Year','Activity Type Code'],aggfunc='sum')

data6.to_csv('Data sorted by year and activity type.csv')
data6
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# data3 = pd.DataFrame(data2)

# data3.reset_index().plot(x="Year", y=["Passenger Count"], kind="bar")

# plt.title("Activity during years")
# plt.xlabel("Years")
# plt.ylabel("Travels count")
```

```python
data7 = data6.pivot_table(index= 'Year', columns='Activity Type Code')
data7
```

```python
data6.pivot_table(index='Year', columns='Activity Type Code').plot(kind='bar')
plt.title("Activity during years")
plt.xlabel("Years")
plt.ylabel("Passengers (mln)")
plt.savefig('Figure1.pdf')
plt.show()
```

```python
data8 = data5.pivot_table(values='Passenger Count',index='Month',aggfunc='sum')
data8.to_csv('Data sorted by month and Passenger count.csv')
data8
```

```python
import pandas as pd
import numpy as np

data9 = data5.pivot_table(values='Passenger Count',index=['Year','Operating Airline','GEO Region'],aggfunc='sum')


data9s = data9.sort_values(['Year', 'Passenger Count'],
              ascending = [True, False]).groupby('Year').head(3)
# data4.groupby('Year').sort_values(by = ['Passenger Count'], ascending=False)
# data91=data9.sort_values(['Passenger Count'],ascending=False).groupby('Year').head(3)
data9s.to_csv('Data sorted by yAirline and Region.csv')
data9s
```

```python
data9s.pivot_table(index='Year', columns='Operating Airline').plot()
plt.title("Average of temp by years")
plt.xlabel("Year")
plt.ylabel("AVG of temp")
plt.show()
```

```python
data10 = data5.loc[data1["GEO Summary"] != 'Domestic']
data10
```

```python
data11 = data10.pivot_table(values='Passenger Count',index=['Year','Operating Airline','GEO Region'],aggfunc='sum')


data12 = data11.sort_values(['Year', 'Passenger Count'],
              ascending = [True, False]).groupby('Year').head(3)
data12.to_csv('Data sorted by International Airline and Region.csv')
data12
```

```python
data12.pivot_table(index='Year', columns='GEO Region').plot()
plt.title("Activity regarding regions")
plt.xlabel("Years")
plt.ylabel("Pasangers (mln)")
plt.savefig('Figure4.pdf')
plt.show()
```
