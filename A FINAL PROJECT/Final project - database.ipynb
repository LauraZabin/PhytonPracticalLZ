```python
import mysql.connector
import pandas as pd
import pymysql

from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:@localhost/air traffic")

df = pd.read_excel(r"C:\Users\pauli\Desktop\PhytonCourse\A FINAL PROJECT\Airline Information.xlsx")

df.to_sql(name="airlines",con=engine,if_exists="append",index=False)
```

```python
from sqlalchemy import create_engine
from sqlalchemy import text

df1 = pd.read_csv(r"C:\Users\pauli\Desktop\PhytonCourse\A FINAL PROJECT\Data transformated.csv")


with engine.begin() as con:
    query = text("SELECT * FROM airlines")
    df2 = pd.read_sql(query,con)
    
Joindf = pd.merge(df1, 
                     df2, 
                     on ='Operating Airline', 
                     how ='inner')
Joindf
```

```python
Joindf.to_csv("passengers statistics.csv",index=False)
```
